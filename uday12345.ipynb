# bar plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,50,90]
z=[20,20,20,20]
plt.xlabel("SUBJECTS",fontsize=30,color="r")
plt.ylabel("NUMBERS",fontsize=30)
plt.title("BAR CHART",fontsize=30,color="g")
plt.bar(x,y,width=0.5,color='r')
plt.bar(x,z,width=0.5,color='y',)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","java"]
y=[85,70,50,90]
w=[30,30,30,40]
z=[20,20,20,20]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
p2=[k+width for k in p]

plt.xlabel("SUBJECTS",fontsize=30,color="r")
plt.ylabel("NUMBERS",fontsize=30)
plt.title("BAR CHART",fontsize=30,color="g")

plt.bar(p,y,width,color='r')
plt.bar(p1,w,width,color='y')
plt.bar(p2,z,width,color='b')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,50,90]
z=[20,20,20,20]
plt.xlabel("SUBJECTS",fontsize=30,color="r")
plt.ylabel("NUMBERS",fontsize=30)
plt.title("BAR CHART",fontsize=30,color="g")
b=["y","m","b","k"]
plt.grid(axis='x',color='r')
plt.grid(axis='y',color='g')
plt.bar(x,y,width=0.5,color=b,align="edge",label="BAR CHAR",edgecolor="r",linewidth=5,linestyle="-.",alpha=1)
plt.bar(x,z,width=0.5,color='y',align="edge",label="BAR CHAR",edgecolor="r",linewidth=5,linestyle="-.",alpha=1)
plt.legend()
plt.show()
```
