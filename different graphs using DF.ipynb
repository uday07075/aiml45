# reading file in dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/UDAY/Desktop/weight-height.csv")
df
print(df.info())
print(df.shape)
print(df.describe)
print(df.size)
print(df.head(5))
print(df.tail(5))
print(df.memory_usage)
#if you want specific column
print(df.at[2,'Gender'])
#if you want 1 to 10 columns in gender
print(df.loc[1:10,'Gender'])
#to get unique value in gender
print(df['Gender'].unique())
```

# to plot  scatter graph

```python
plt.scatter(x=df['Height'],y=df['Weight'],marker='d')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('height and weights in gender')
#to plot x axis grid 
plt.grid('x axis')
plt.show()
```

```python
x=df['Gender']
y=df['Weight']
plt.xlabel('Gender',family='magneto',color='b',size=30)
plt.ylabel('Weight',family='magneto',color='y',size=30)
plt.bar(x,y,color='r',linewidth=1,edgecolor='g',linestyle='-.')
plt.title('Gender weights',loc='left',family='magneto',color='g',size=50)
plt.show()
```

```python
#to plot histogram using a dataframe
plt.hist(df['Height'],bins=12)
plt.title('histogram')
plt.show()
```
